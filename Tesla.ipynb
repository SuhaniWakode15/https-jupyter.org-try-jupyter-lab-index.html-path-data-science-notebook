```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page you want to scrape
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Define headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

# Send an HTTP request to the website with the headers
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Print first 1000 characters of the page to check if the table exists
    print(response.text[:1000])
    
    # Try to find all tables on the page
    tables = soup.find_all('table')

    # Check if any tables are found
    if tables:
        # Extract the first table
        revenue_table = tables[0]  # Assuming the first table contains the revenue data

        # Convert the table to a DataFrame
        tesla_revenue = pd.read_html(str(revenue_table))[0]

        # Display the first few rows of the dataframe
        print(tesla_revenue.head())
    else:
        print("No tables found on the page.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
# Extract GameStop stock data
gme_data = yf.download('GME', period='1y')

# Reset index and display first 5 rows
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL with GameStop's revenue data
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Set headers to simulate a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Send the GET request with headers
response = requests.get(url, headers=headers)

# Check if the page was retrieved successfully
if response.status_code == 200:
    # Parse the page content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Print first 1000 characters to inspect the page content
    print(response.text[:1000])

    # Try to find all tables on the page
    tables = soup.find_all('table')

    # Check if any tables are found
    if tables:
        # Extract the first table
        revenue_table = tables[0]  # Assuming the first table contains the revenue data

        # Convert the table to a DataFrame
        gme_revenue = pd.read_html(str(revenue_table))[0]

        # Display the last 5 rows of the dataframe
        print(gme_revenue.tail())
    else:
        print("No tables found on the page.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
